Box Plot and outlier detection from plot
---------

```python
a = [22, 44, 66, 88, 99, 55, -250, 220, 300, 500, 1000, 1200]
fig = plt.figure(figsize=(15,5))
green_diamond = dict(markerfacecolor='r', marker='D')
plt.boxplot(a, flierprops=green_diamond)
plt.grid()
plt.xlabel('x values as length of y ')
plt.ylabel('a list values')
# plt.legend(loc='upper left')
plt.title('simple plot for one set of data')
plt.show()
fig.savefig("box_plot.png")
print("________________________________________________________")
from numpy import percentile
# calculate interquartile range
q25, q75 = percentile(a, 25), percentile(a, 75)
iqr = q75 - q25
print('Percentiles: 25th=%.3f, 75th=%.3f, IQR=%.3f' % (q25, q75, iqr))
# calculate the outlier cutoff
cut_off = iqr * 1.5
lower, upper = q25 - cut_off, q75 + cut_off
print(lower)
print(upper)
# identify outliers
# outliers = [x for x in a if x < lower or x > upper]
for x in a:
    if x < lower or x > upper:
        print(x)
# print('Identified outliers: %d' % len(outliers))
# remove outliers
outliers_removed = [x for x in a if x >= lower and x <= upper]
print('Non-outlier observations: %d' % len(outliers_removed))
```
