# Decision Tree (Supervised Learning) - Classification & Regression 
--------

Decision Tree algorithm belongs to the family of supervised learning algorithms

Decision tree algorithm can be used for solving regression and classification problems

The general aim of using Decision Tree is to create a training model which we can use to predict class or value of target variables by learning decision rules inferred from training data

Tree = Root Node + Leaf Node

Leaf node is decision or classification 

Top most decision node is root node

Decision trees can handle both categorical and numerical data

Decision tree is top-down from a root node and involves partioning of data into subsets that contain instances with similar values 

Decision tree - classification
-----

Decision tree uses 2 types of classification criteria's : “gini” criteria for Gini Index & “entropy” for Information Gain. By default, it takes “gini” value

Understanding Entropy : 

1. Entropy ranges between 0 and 1 

2. Entropy = 0 means sample is completely homogeneous ---> Leaf Node


3. Entropy = 1 means sample is Equally divided --> Impure set , Good for training

```python
import pandas as pd
data = pd.read_csv("decisiontree_data.csv", header= None)
data.head(3)
# data.shape
# 4 input columns and one ouptut column(multinomial classification)
```

```python
X = data.values[:, 1:5]
Y = data.values[:,0]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size = 0.25, random_state = 100)

# Random state is used to keep the data splitting static every time with static coefficient as 18 
# Test_size + Train_size = 1 
```

```python
import sklearn.tree
dir(sklearn.tree)
```

```python
from sklearn.tree import DecisionTreeClassifier # importing the required classification class from sklearn 

clf_gini = DecisionTreeClassifier(criterion = "gini") # creating instance of class 

#clf_gini = DecisionTreeClassifier(criterion = "gini", random_state = 100, max_depth=3, min_samples_leaf=5)

# criterion = gini or entropy , default is gini 

# splitter = best or random , default is best 

# max_depth = integer value , default is none 
```

```python
dir(clf_gini)
```

```python
clf_gini.fit(X_train, y_train) # building your model for classification with gini as criterion on testing data 

# X_train , y_train is all your training data 
```

```python
from sklearn.tree import export_graphviz

export_graphviz(clf_gini,'clf_gini.tree')

# The clf_gini.tree is stored in your project directory 
```

```python
# Go to this link   http://www.webgraphviz.com/   and see the graph by applying tree content and click on generate graph 

# From tree you can see the lead node with gini value as 0
```

```python
print("gini index criteria decision tree is completed and is shown below")
from IPython.display import Image
Image(filename = "clf_gini_tree.png",width=1000,height=1000)
```

```python
clf_gini.predict([[4, 1, 1, 3]])
```

```python
y_pred_gini = clf_gini.predict(X_test)
```

```python
y_pred_gini
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred_gini)*100
```

```python
clf_entropy = DecisionTreeClassifier(criterion = "entropy")

clf_entropy.fit(X_train, y_train)

export_graphviz(clf_entropy,'clf_entropy.tree')

print(clf_entropy.predict([[4, 1, 1, 3]]))

y_pred_entropy = clf_entropy.predict(X_test)

accuracy_score(y_test,y_pred_entropy)*100

```

```python
print("Entropy criteria decision tree is completed and is shown below")
from IPython.display import Image
Image(filename = "clf_entropy_tree.png",width=1000,height=1000)
```

```python
from sklearn.metrics import confusion_matrix
cm_gini = confusion_matrix(y_test, y_pred_gini)
cm_entropy = confusion_matrix(y_test, y_pred_entropy)
```

Summary of Decision tree classifier :
----


1 . Import the data(csv file) and create data frame 


2. After creating data frame , create input and output dataframes 

3. split the input and output dataframes into testing and training data 

4. import the required decision tree classifier class and create instance 

5 .After creating instance apply fit on training data

6. After buiding your model apply predict on testing input data and see the predicted output data 

7 . conclude tradeoff between your testing data and model predicted data with help of sklearn classification metrics like accuracy_score

8 . see the confusion matrix and understand the diagonal elements are higher in weight for accurate predictions 

9 . If required calculate the precision , recall, etc from your confusion matrix 

10 . same above process steps 1 - 9 repeat with instance of criteria called as entropy


11 . Now understand that always gini index is much faster and good to use than entropy , enrtopy involves logarthimic calculations where as gini doesnt involves log2

similarly Decision tree regression uses mean squared error to split/decide the nodes
