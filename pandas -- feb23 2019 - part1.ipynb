```python
# use fillna() to fill the NaN column with values 
# Below we are filling with 1000 value at a time for 2 columns Volume and opencloseavg
yesbank[['opencloseavg','Volume']] = yesbank[['opencloseavg','Volume']].fillna(1000)
yesbank[['Open']] = yesbank[['Open']].fillna(1000) # filling NaN with 1000 for single column 
yesbank
```

```python
yesbank[['Low']] = yesbank[['Low']].fillna(100000)
yesbank[['High']] = yesbank[['High']].fillna(method = 'bfill')
yesbank
# bfill method takes the back element of NaN and fills 
# similarly ffill method takes forward element of NaN and fills 
```
