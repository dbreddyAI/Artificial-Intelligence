Dataframes Merging
--------

Two dataframes can be merged based 2 parameters 

on = column name 
how = inner , outer , left , right 

```python
import pandas as pd
import numpy as np

# Another way of creating dataframes 
 
# data frame 1
d1 = {'Customer_id':pd.Series([1,2,3,4,5,6]),'st_id':pd.Series([10,20,30,40,50,60]),
  'Product':pd.Series(['p1','p2','p3','p4','p5','p6'])}
df1 = pd.DataFrame(d1)
 
 
# data frame 2
d2 = {'Customer_id':pd.Series([2,4,6]),
    'State':pd.Series(['ap','tn','ka'])}
df2 = pd.DataFrame(d2)
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on='Customer_id', how='inner') # Inner Join
```

```python
pd.merge(df1, df2, on='Customer_id', how='outer') # Outer join 
```

```python
pd.merge(df1, df2, on='Customer_id', how='left') # Left outer join 
```

```python
pd.merge(df1, df2, on='Customer_id', how='right') # Right outer join
```

```python
import numpy as np
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data1': range(6),
                   'data2': range(0,11,2)},
                   columns = ['key', 'data1', 'data2'])
```

```python
df
```

```python
dfgrp = df.groupby('key')
```

```python
dfgrp
```

```python
dfgrp.groups
```

```python
dfgrp.size()
```

```python
dfgrp.aggregate(['min',max, np.median])
```
